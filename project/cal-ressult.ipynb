```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import sys
import os
```

```python
df=pd.read_csv(os.path.join(os.getcwd(),'..\data\\final.csv'),index_col=0)
df
```

```python
covrdf=df[['AvgTemperature','PricePerMiles']].cov()
matrix = np.triu(covrdf)

# using the upper triangle matrix as mask 
fig, ax = plt.subplots(figsize=(12,8))
sns.heatmap(covrdf, ax=ax, fmt='.1', annot=True, mask=matrix)
plt.show()
```

```python
fig, ax1 = plt.subplots()
color = 'tab:blue'
ax1.set_ylabel('Price Per Miles', color=color)
ax1.set_xlabel('Quarter',)
ax1.plot(df['quarter'], df['PricePerMiles'], color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.xaxis.set_major_locator(plt.MaxNLocator(integer=True))

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Avg Temperature', color=color)
ax2.tick_params(axis='y', labelcolor=color)

ax2.plot(df['quarter'], df['AvgTemperature'], color=color)
fig.tight_layout()
plt.show()
```
